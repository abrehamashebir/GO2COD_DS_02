### Model  Selection, Training and Pridiction

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.model_selection import train_test_split
# 1. Import necessary libraries
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE
```

```python
# Selecting Feature and target variable
X = card.drop('Class', axis=1)
y = card['Class']
```

```python
# Scalling and Transfofrming Data

scalar = StandardScaler()
X_1 = scalar.fit_transform(X)
card_scaled = pd.DataFrame(X_1, columns=X.columns)
x_train, x_test, y_train, y_test = train_test_split(card_scaled, y , test_size=0.20, random_state=42)
```

```python
sns.countplot(x='Class', data=card)
plt.title(' Distribution')
plt.show()
```

```python
y_train.value_counts().plot(kind='bar')
```

```python
# Handle Class Inbalance
smote = SMOTE(random_state=42)
x_train_trans, y_train_trans =smote.fit_resample(x_train, y_train)
```

```python
y_train_trans.value_counts().plot(kind='bar')
```

```python
# Using LLogistic Regression

model_lr = LogisticRegression(random_state=42)
model_lr.fit(x_train_trans, y_train_trans)
y_pred_lr = model_lr.predict(x_test)
```

```python
cmr_lr = confusion_matrix(y_test, y_pred_lr)
```

```python
print(cmr_lr)
```

```python
plt.imshow(cmr_lr, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix - Logistic Regression')
plt.colorbar()
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
print(classification_report(y_test, y_pred_lr))
```
